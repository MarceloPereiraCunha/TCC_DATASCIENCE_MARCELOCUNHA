```python

#FONTE: https://www.rs-online.com/designspark/predicting-weather-using-lstm
#


# Core Keras libraries
#
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Bidirectional
 
#
# For data conditioning
#
from scipy.ndimage import gaussian_filter1d
from scipy.signal import medfilt

#
# Make results reproducible
#
from numpy.random import seed
seed(1)
#from tensorflow import set_random_seed
#set_random_seed(1)

import tensorflow
tensorflow.random.set_seed(1)


# 
# Other essential libraries
#
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from numpy import array

# Make our plot a bit formal
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 10}
plt.rc('font', **font)


#
# Set input number of timestamps and training days
#
#default n_timestamp = 10
n_timestamp = 24
#train_days = 1500  # number of days to train from
#testing_days = 500 # number of days to be predicted
train_days = 4000  # number of days to train from
testing_days = 2391 # number of days to be predicted
validation_days = 2390 # number of days to be predicted
n_epochs = 25
filter_on = 1
#8781

#
# Select model type
# 1: Single cell
# 2: Stacked
# 3: Bidirectional
#

model_type = 2

path='/home/marcelo/Documentos/PosGraduacaoDataScience/Disciplinas/Monografia/DadosMonografia/dados_inmet_filtrados_2020.csv'
dataset = pd.read_csv(path)
dataset = dataset.drop(['Hora UTC'], axis=1)

#url = "https://raw.githubusercontent.com/danrustia11/WeatherLSTM/master/data/weather_temperature_yilan.csv"
#dataset = pd.read_csv(url)
#if filter_on == 1:
    #dataset['Temperature'] = medfilt(dataset['Temperature'], 3)
    #dataset['Temperature'] = gaussian_filter1d(dataset['Temperature'], 1.2)

if filter_on == 1:
    dataset['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (Celsius)'] = medfilt(dataset['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (Celsius)'], 3)
    dataset['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (Celsius)'] = gaussian_filter1d(dataset['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (Celsius)'], 1.2)


#
# Set number of training and testing data
# 
train_set = dataset[0:train_days].reset_index(drop=True)
test_set = dataset[train_days: train_days+testing_days].reset_index(drop=True)
validation_set = dataset[testing_days:8781].reset_index(drop=True)
training_set = train_set.iloc[:, 1:2].values
testing_set = test_set.iloc[:, 1:2].values
validation_set = validation_set.iloc[:, 1:2].values



#
# Normalize data first
#
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)
testing_set_scaled = sc.fit_transform(testing_set)
validating_set_scaled=sc.fit_transform(validation_set)



#
# Split data into n_timestamp
#

def data_split(sequence, n_timestamp):
    X = []
    y = []
    for i in range(len(sequence)):
        end_ix = i + n_timestamp
        if end_ix > len(sequence)-1:
            break
        # i to end_ix as input
        # end_ix as target output
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)


X_train, y_train = data_split(training_set_scaled, n_timestamp)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test, y_test = data_split(testing_set_scaled, n_timestamp)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

X_validation, y_validation = data_split(validating_set_scaled, n_timestamp)
X_validation = X_validation.reshape(X_validation.shape[0], X_validation.shape[1], 1)


if model_type == 1:
    # Single cell LSTM
    model = Sequential()
    model.add(LSTM(units = 50, activation='relu',input_shape = (X_train.shape[1], 1)))
    model.add(Dense(units = 1))
if model_type == 2:
    # Stacked LSTM
    model = Sequential()
    model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(LSTM(50, activation='relu'))
    model.add(Dense(1))
if model_type == 3:
    # Bidirectional LSTM
    model = Sequential()
    model.add(Bidirectional(LSTM(50, activation='relu'), input_shape=(X_train.shape[1], 1)))
    model.add(Dense(1))





#
# Start training
#
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
history = model.fit(X_train, y_train, epochs = n_epochs, batch_size = 32)
loss = history.history['loss']
epochs = range(len(loss))






#
# Get predicted data
#
y_predicted = model.predict(X_test)


#
# Validating The Model
#
y_predicted_validating = model.predict(X_validation)



#
# 'De-normalize' the data
#
y_predicted_validating_descaled = sc.inverse_transform(y_predicted_validating)
y_predicted_descaled = sc.inverse_transform(y_predicted)
y_train_descaled = sc.inverse_transform(y_train)
y_test_descaled = sc.inverse_transform(y_test)
y_validation_descaled = sc.inverse_transform(y_validation)
y_pred = y_predicted.ravel()
y_pred = [round(yx, 2) for yx in y_pred]
y_tested = y_test.ravel()
y_validated = y_validation.ravel()







#
# Show results
#
plt.figure(figsize=(16,14))

plt.subplot(4, 1, 1)
plt.plot(dataset['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (Celsius)'], color = 'black', linewidth=1, label = 'True value')
plt.ylabel("Umidade Relativa (h)")
plt.xlabel("Hora")
plt.title("Toda Srie Temporal")


plt.subplot(4, 2, 3)
plt.plot(y_test_descaled, color = 'black', linewidth=1, label = 'Valor Real')
plt.plot(y_predicted_descaled, color = 'red',  linewidth=1, label = 'Valor Previsto')
plt.legend(frameon=False)
plt.ylabel("Umidade Relativa (h)")
plt.xlabel("Hora")
plt.title("Previso (n Horas)")

plt.subplot(4, 2, 4)
plt.plot(y_test_descaled[0:75], color = 'black', linewidth=1, label = 'True value')
plt.plot(y_predicted_descaled[0:75], color = 'red', label = 'Previsto')
plt.legend(frameon=False)
plt.ylabel("Umidade Relativa (h)")
plt.xlabel("Hora")
plt.title("Dados Previstos (Primeiras 75 horas)")

plt.subplot(4, 3, 7)
plt.plot(epochs, loss, color='black')
plt.ylabel("Perda (MSE)")
plt.xlabel("poca")
plt.title("Curva de Treinamento")

plt.subplot(4, 3, 8)
plt.plot(y_test_descaled-y_predicted_descaled, color='black')
plt.ylabel("Residuo")
plt.xlabel("Hora")
plt.title("Curva de Resduo")

plt.subplot(4, 2, 7)
plt.scatter(y_predicted_descaled, y_test_descaled, s=2, color='black')
plt.ylabel("Y Real")
plt.xlabel("Y Previsto")
plt.title("Scatter plot")

plt.subplot(4, 2, 8)
plt.scatter(y_predicted_validating_descaled, y_validation_descaled, s=2, color='black')
plt.ylabel("Y Real")
plt.xlabel("Y Validado")
plt.title("Scatter plot")

plt.subplots_adjust(hspace = 0.5, wspace=0.3)
plt.show()



mse_Test = mean_squared_error(y_test_descaled, y_predicted_descaled)
mse_Validation = mean_squared_error(y_validation_descaled, y_predicted_validating_descaled)

r2_Test = r2_score(y_test_descaled, y_predicted_descaled)
r2_Validation = r2_score(y_validation_descaled, y_predicted_validating_descaled)

print("Mean Square Error on Testing (mse) = " + str(round(mse_Test,9)))
print("Mean Square Error on Validating (mse) = " + str(round(mse_Validation,9)))
print("r2 Test =" + str(round(r2_Test,9)))
print("r2= Validation =" + str(round(r2_Validation,9)))
```

```python
class WindowGenerator():
  def __init__(self, input_width, label_width, shift,
               train_df=train_df, val_df=val_df, test_df=test_df,
               label_columns=None):
    # Store the raw data.
    self.train_df = train_df
    self.val_df = val_df
    self.test_df = test_df

    # Work out the label column indices.
    self.label_columns = label_columns
    if label_columns is not None:
      self.label_columns_indices = {name: i for i, name in
                                    enumerate(label_columns)}
    self.column_indices = {name: i for i, name in
                           enumerate(train_df.columns)}

    # Work out the window parameters.
    self.input_width = input_width
    self.label_width = label_width
    self.shift = shift

    self.total_window_size = input_width + shift

    self.input_slice = slice(0, input_width)
    self.input_indices = np.arange(self.total_window_size)[self.input_slice]

    self.label_start = self.total_window_size - self.label_width
    self.labels_slice = slice(self.label_start, None)
    self.label_indices = np.arange(self.total_window_size)[self.labels_slice]

  def __repr__(self):
    return '\n'.join([
        f'Total window size: {self.total_window_size}',
        f'Input indices: {self.input_indices}',
        f'Label indices: {self.label_indices}',
        f'Label column name(s): {self.label_columns}'])
```
